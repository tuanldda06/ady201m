```python
df = df[df["type_single"] == True]
```

```python
print(len(df))
```

```python
df["lot_living"] = df["living"]
```

```python
cols_to_drop = [
    # target cũ nếu không dùng
    "type_condo",
    "type_manufactured",
    "type_multi",
    "type_single",
    "type_townhouse",

    # dễ gây trùng lặp thông tin
    "pop",
    "risk_overall",
    "bed_bath",
    "risk_loss",
    "risk_social",
    "risk_resilience",
    "living"
    
]

df = df.drop(columns=cols_to_drop)
```

```python
print(len(df[df["bed"] == 0]))
print(len(df[df["bath"] == 0]))
print(len(df[df["lot_sqft"] == 0]))
print(len(df[df["lot_living"] == 0]))
print(len(df[df["price"] == 0]))
```

```python
import numpy as np
df.loc[df["lot_sqft"] == 0, "lot_sqft"] = np.nan
df = df.dropna(
    subset=[
        "lot_sqft"
    ]
)
```

```python
df.nsmallest(
    10,
    "lot_living"
)[[ "lot_sqft", "lot_living"]]
```

```python
df = df[df["lot_sqft"] < 1_000_000]
```

```python
print(len(df))
```
