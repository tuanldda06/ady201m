# Prepare Final Dataset

**Input:** `data_csv/zillow_clean.csv`

**Output:** `data_csv/zillow_final.csv`

Mục tiêu của notebook này là tạo dataset cuối cùng cho EDA và Modeling: lọc dòng không hợp lệ, xử lý missing values, tính lại các biến tỷ lệ, bỏ loại nhà `type_lot`, xử lý outlier giá cao và lưu `zillow_final.csv`.

```python
import sys
assert sys.version_info >= (3, 5)

import warnings
warnings.filterwarnings('ignore')

import sklearn
assert sklearn.__version__ >= "0.20"

import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import math

np.random.seed(21)
```

```python
df = pd.read_csv('../data_csv/zillow_clean.csv')
```

```python
df.dropna(subset=['price'], inplace=True)
```

```python
df = df.dropna(subset=["price"])
df = df[df["price"] != 0].copy()
```

```python
#Bỏ những dòng là đất / không có thông tin nhà
df = df.dropna(
    subset=["bed", "bath", "living"],
    how="all"
)
# Bỏ dòng bed = 0, bath = 0, living = 0

bad_rows = (
    (df["bed"].fillna(0) == 0) &
    (df["bath"].fillna(0) == 0) &
    (df["living"].fillna(0) == 0)
)

df = df[~bad_rows].copy()
```

Handling missing values

```python
def missingData(data):
    total = data.isnull().sum().sort_values(ascending = False)
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_values(ascending = False)
    md = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    md = md[md["Percent"] > 0]
    plt.figure(figsize = (20, 8))
    plt.barh(md.index, md["Percent"],color="g",alpha=0.8)
    plt.title('Percent missing data by feature', fontsize=15)
    return md
missingData(df)
```

Ta thấy bed , living , bath missing không quá nhiều -> Bỏ để mô hình ổn định , sạch


```python
df = df.dropna(subset=["bed", "bath", "living"])
df = df[
    (df["bed"] > 0) &
    (df["bath"] > 0) &
    (df["living"] > 0)
].copy()
```

Còn những giá trị còn lại thì chúng ta sẽ dùng median ( lấy trung vị) để xử lý 

```python
median_cols = [
    "rent",
    "income",
    "lot_sqft",
    "unemployment",
    "owner_occ",
    "bachelor",
    "poverty",
    "pop",
    "pop_density",
    "bed",
    "bath",
    "living"
]
median_cols = [col for col in median_cols if col in df.columns]

for col in median_cols:
    df[col] = df[col].fillna(df[col].median())


df["bed_bath"] = df["bed"] / df["bath"].replace(0, np.nan)

df["lot_living"] = df["lot_sqft"] / df["living"].replace(0, np.nan)

# Nếu sau khi chia vẫn còn NaN thì fill median nốt
df["bed_bath"] = df["bed_bath"].fillna(df["bed_bath"].median())
df["lot_living"] = df["lot_living"].fillna(df["lot_living"].median())
```

```python
print("Sau xử lý missing/invalid rows, dữ liệu đã sạch missing cho các biến mô hình.")
print("Bước tiếp theo: bỏ cột phụ type_lot và lọc outlier giá trước khi lưu final dataset.")
display(df.head())
```

```python
df = df.drop(columns=["type_lot"], errors="ignore") # Bỏ loại đất trống vì không còn dòng hợp lệ cho mô hình
```

```python
before_outlier = len(df)
upper = df["price"].quantile(0.95)
df = df[df["price"] <= upper].copy()

print(f"Ngưỡng 95th percentile của price: {upper:,.0f}")
print(f"Đã loại {before_outlier - len(df)} dòng giá cao nhất/cực đoan")
print(f"Số dòng sau outlier filtering: {len(df):,}")
```

```python
df.to_csv("../data_csv/zillow_final.csv", index=False, encoding="utf-8-sig")

print(f"Saved zillow_final.csv with shape: {df.shape}")
print("Total missing:", df.isna().sum().sum())
print("Duplicate rows:", df.duplicated().sum())
print(df.columns.tolist())
```
